```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load dataset
housing_data = fetch_california_housing(as_frame=True)

# Convert to DataFrame
housing = housing_data.frame

# Add Ocean Proximity column (for lab requirement)
housing['ocean_proximity'] = 'NEAR OCEAN'

# Save as CSV
housing.to_csv("housing.csv", index=False)

housing.head()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load dataset
diabetes = pd.read_csv("/content/diabetes.csv")

# Display first 5 rows
print("First 5 rows:")
display(diabetes.head())

# Check missing values
print("\nMissing Values Before Handling:")
print(diabetes.isnull().sum())

# Handle missing values
diabetes.fillna(diabetes.mean(), inplace=True)

# Handle outliers using IQR
Q1 = diabetes.quantile(0.25)
Q3 = diabetes.quantile(0.75)
IQR = Q3 - Q1

diabetes = diabetes[~((diabetes < (Q1 - 1.5 * IQR)) |
                      (diabetes > (Q3 + 1.5 * IQR))).any(axis=1)]

# Min-Max Scaling
minmax = MinMaxScaler()
diabetes_minmax = minmax.fit_transform(diabetes)

# Standard Scaling
standard = StandardScaler()
diabetes_standard = standard.fit_transform(diabetes)

print("\nDiabetes Dataset Shape After Preprocessing:")
print(diabetes.shape)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load dataset
adult = pd.read_csv("/content/adult.csv")

# Display first 5 rows
print("First 5 rows:")
display(adult.head())

# Check missing values
print("\nMissing Values Before Handling:")
print(adult.isnull().sum())

# Handle missing values
adult.replace('?', np.nan, inplace=True)
adult.fillna(adult.mode().iloc[0], inplace=True)

# Identify categorical columns
categorical_cols = adult.select_dtypes(include='object').columns
print("\nCategorical Columns:")
print(categorical_cols)

# Encode categorical columns
le = LabelEncoder()
for col in categorical_cols:
    adult[col] = le.fit_transform(adult[col])

# Standard Scaling
scaler = StandardScaler()
adult_scaled = scaler.fit_transform(adult)

print("\nAdult Dataset Shape After Preprocessing:")
print(adult.shape)
```
